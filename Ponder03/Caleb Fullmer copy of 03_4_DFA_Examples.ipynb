```python
odd0Even1 = md2mc('''
DFA

!! DFA that accepts so long as the string seen so far 
!! has odd0s and even1s (in any order)

!! State design: I'll tag along letters "EO, OE, etc."
!! to indicate parity of 0's and 1's

I_EE : 0 -> F_OE
I_EE : 1 -> S_EO

S_EO : 0 -> S_OO
S_EO : 1 -> I_EE

F_OE : 0 -> I_EE
F_OE : 1 -> S_OO

S_OO : 0 -> S_EO
S_OO : 1 -> F_OE

''')
```

```python
dotObj_dfa(odd0Even1)
```

```python
# Now arrive at this via DFA intersection

Odd0 = md2mc('''
DFA
I : 1 -> I
I : 0 -> F

F : 1 -> F
F : 0 -> I
''')

Even1 = md2mc('''
DFA
IF : 0 -> IF
IF : 1 -> S

S : 0 -> S
S : 1 -> IF
''')

Odd0_and_Even1 = intersect_dfa(Odd0, Even1)
```

```python
dotObj_dfa(Odd0)
```

```python
dotObj_dfa(Even1)
```

```python
dotObj_dfa(Odd0_and_Even1)
```

```python
# DeMorgan's law:
print(comp_dfa(union_dfa(Odd0, Even1)) == intersect_dfa(comp_dfa(Odd0), comp_dfa(Even1)))
print(comp_dfa(intersect_dfa(Odd0, Even1)) == union_dfa(comp_dfa(Odd0), comp_dfa(Even1)))
```

```python
Odd0FbyEven1 = md2mc('''
DFA

!! DFA that accepts so long as the string seen so far 
!! has odd0s and even1s (in THAT order)

!! State design: Must close-off 0's after seeing a 1
!! I'll denote this by adding a 'd' for the 0 position
!! So S_EE, Sd_EE, Fd_EE etc, where the 'd' says 0's are unwelcome

I_EE : 0 -> F_OE
I_EE : 1 -> Sd_EO

Sd_EO : 0 -> BH
Sd_EO : 1 -> Sd_EE

F_OE : 0 -> I_EE
F_OE : 1 -> Sd_OO

Sd_OO : 0 -> BH
Sd_OO : 1 -> Fd_OE

Sd_EE : 0 -> BH
Sd_EE : 1 -> Sd_EO

Fd_OE : 0 -> BH
Fd_OE : 1 -> Sd_OO

BH  : 0|1 -> BH
''')
```

```python
dotObj_dfa_w_bh(Odd0FbyEven1)
```
