With the unary, so long as the top and bottom are not identical for two or more tiles, and there are at least two tiles where the top of tile1 has fewer character than the bottom of tile1 and the top of tile2 has more characters than the bottom of tile2, then at least one solution exists. However, this guarantee does not exist for binary.
