1. Describe a decider to check whether the language of a given DFA $D$ is infinite.

Question 1 Answer:

As discussed in our group, a decider for this could be created by checking for the star operator in the DFA. If the star operator exists, then the language must be infinite. However, two issues can occur with this. One, it would be difficult to determine loops that are not dependant on a star. So for example, if one state loops back to another. Second, if the star operator occurs on a black-hole state, we could get a false positive.

To fix this, we will reduce our problem of finding if the language of a DFA is infinite to the equivalent problem of determining if a regular expression contains the star operator. This is done by converting the DFA to an NFA and then converting the NFA to a regular expression. This should remove the issue of black-hole states and make finding loops such as (abc)* much easier.

2. Describe a decider to check whether a linear bounded automaton $L$ working on an input $w$ will go into an infinite loop.

Question 2 Answer:

This is just the halting problem. It would output “yes halted” if the machine goes into an infinite loop or “does not halt” if it does go into an infinite loop. However, we know this is undecidable. So, we cannot create any such decider.
